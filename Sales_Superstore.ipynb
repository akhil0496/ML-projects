# Data Preprocessing and Model building

```python
df.head()
```

By seeing above data we can remove the product Id ,city,Row id,order id,order data,ship date,ship mode,customer id,customer name,segment,product name.

```python
df_filtered=df[['Region','Sub-Category','Sales','Quantity','Discount','Profit','Category','ProductName']]
df_filtered.head() #getting important columns from data
```

Given Features does not say much. so do feature engineering, it helps to build better model.

```python
#feature engineering
df_filtered['Sales_per_piece']=df_filtered['Sales']/df_filtered['Quantity'] #Sales per each item
```

```python
df_filtered.isnull().sum() #Checking missing values
```

```python
import seaborn as sns
sns.distplot(df_filtered['Sales'])
print('Skweness: %f' % df_filtered['Sales'].skew())
print('Kurtsis: %f' %df_filtered['Sales'].kurt()) #The sales(target variable) are not distributed normally we can see it is right skew 
```

```python
df_filtered['Sales']=np.log(df_filtered["Sales"])#Do log transformation to get normal distribution
sns.distplot(df_filtered['Sales'])
```

```python
df_filtered['Discount']=df_filtered['Discount'].apply(lambda x: 'WithOut Discount' if x==0 else 'With Discount') #Converting to with or without discount
```

```python
#Categorical features
categorical_features=df_filtered.select_dtypes(include=[np.object])
categorical_features.head()
```

```python
#Numerical Features
numerical_features=df_filtered.select_dtypes(include=[np.number])
numerical_features.head()
```

```python
for col in categorical_features:
    print('\n%s column:'%col)
    print(df_filtered[col].value_counts())
```

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in categorical_features:
    df_filtered[i]=le.fit_transform(df_filtered[i])
```

```python
df_filtered.head()
```

```python
#We can find outliers by boxplots
df_filtered.boxplot()
plt.xticks(list(range(len(df_filtered.columns))),df_filtered.columns,rotation='vertical')
```

```python
# Relationship of Category and Sales
df_filtered['Category'].value_counts()
sns.boxplot( "Category","Sales",data= df_filtered )
```

```python
#Remove outliers by IQR
Q1 = df_filtered.quantile(0.20)
Q3 = df_filtered.quantile(0.80)
IQR = Q3 - Q1
print(IQR)
```

```python
data_out= df_filtered[~((df_filtered <(Q1 - 1.5*IQR)) | (df_filtered > (Q3 +1.5 *IQR))).any(axis=1)]
```

```python
#Rechecking outliers by boxplots
#plt.rcParams['figure.figsize'] = (12, 8)
#sns.boxplot( data=data_out)
```

```python
data_out.head()
```

```python
#Corelation between variables
corr=data_out.corr()
sns.heatmap(corr)
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
scl=StandardScaler()
data_out[['Profit','ProductName','Sales_per_piece']]=scl.fit_transform(data_out[['Profit','ProductName','Sales_per_piece']])
```

```python
#One hot encoding
data_df=pd.get_dummies(data_out,columns=['Sub-Category','Region','Quantity','Discount','Category'])
data_df.head()
```

```python
#data_df.dtypes #Dtypes of variables
```

```python
#Data ready for model buliding
x=data_df.drop(['Sales'],axis=1)#input data
y=data_df['Sales']#target variable
```

```python
print(x.shape)
print(y.shape)
```

```python
#Splitting the data for training and validation 
from sklearn.model_selection import train_test_split
X_train,X_val,y_train,y_val=train_test_split(x,y,random_state=0)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression(fit_intercept=True,n_jobs=5,normalize=True)
model.fit(X_train,y_train)
```

```python
print('R-Squared score (training):{:.3f}'.format(model.score(X_train,y_train)))
print('R-Squared score (validation):{:.3f}'.format(model.score(X_val,y_val)))
```

# Ridge Regression

```python
from sklearn.linear_model import Ridge
lr = Ridge(alpha=10.0, fit_intercept=True)
lr.fit(X_train,y_train)
```

```python
print('R-Squared score Ridge (training):{:.3f}'.format(lr.score(X_train,y_train)))
print('R-Squared score Ridge (validation):{:.3f}'.format(lr.score(X_val,y_val)))
```

# RandomForest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
rf= RandomForestRegressor(n_estimators=20,max_features=25,n_jobs=20,verbose=0,random_state=50)
rf.fit(X_train,y_train)
```

```python
print('R-Squared score RandomForest (training):{:.3f}'.format(rf.score(X_train,y_train)))
print('R-Squared score RandomForest (validation):{:.3f}'.format(rf.score(X_val,y_val)))
```

```python
print(y_train.head())
print(y_train.dtype)
```

```python
train=np.exp(y_train)#Convering log(sales) data to normal form by using exponatial form
val=np.exp(y_val)
print(train.shape)
train.head()
```

```python
pred=rf.predict(X_train)
pred2=rf.predict(X_val)
print(pred)
pd.DataFrame(pred2)
```

```python
#converting log(prediction) to normal form
prediction= np.exp(pred)
val_prediction=np.exp(pred2)
```

```python
plt.style.use('fivethirtyeight')
plt.scatter(prediction,prediction-train,
           color='green',s=10,label='Train data')
plt.scatter(val_prediction,val_prediction-val,
           color='blue',s=10,label='Val data')
plt.hlines(y=0,xmin=100,xmax=4000,linewidth=2)
plt.legend(loc='upper right')
plt.title('Residual errors')
plt.show()
```

```python
def mape(y, y_hat):
    err = np.mean(abs((y-y_hat)/y))
    return(err)

def rmse(y, y_hat):
    err = np.sqrt(np.mean(np.square(y-y_hat)))
    return(err)
```

```python
train_mape_3 = mape(train, y_hat=prediction)
val_mape_3 = mape(val, y_hat=val_prediction)
```

```python
print(train_mape_3)
print(val_mape_3)
```

```python
train_rmse=rmse(train,y_hat=prediction)
val_rmse=rmse(val,y_hat=val_prediction)
```

```python
print('Root mean Squared error(training):',train_rmse)
print('Root mean Squared error(validation):',val_rmse)
```
